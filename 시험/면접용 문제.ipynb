### 1. 아래의 문제를 Python으로 풀어주세요.

* 문제 설명 : 재민이는 정수 하나가 적혀있는 숫자 카드 N개를 갖고 있다. 정수 M개가 주어졌을 때, 이 수가 적혀있는 숫자 카드를 재민이가 몇 개 가지고 있는지 구하라.
  
  
* 입력 설명 : 첫 줄에는 재민이가 갖고 있는 카드의 개수 N이 주어진다.
둘째 줄에는 숫자 카드에 적혀있는 정수가 순서대로 주어진다.
셋째 줄에는 M이 주어진다.
넷째 줄에는 재민이가 몇 개 가지고 있는 숫자 카드인지 구해야 할 M개의 정수가 주어진다.
  
  
* 출력 설명 : 첫째 줄에 입력으로 주어진 M개의 수에 대해서 각 수가 적힌 숫자 카드를 재민이가 몇 개 가지고 있는지 공백으로 구분해 출력한다.
  
  
* 입력 예시 :
  
6
  
4 9 1 1 -5 10
  
4
  
1 2 3 -5


* 출력 예시 : 
  
3 0 0 1 2 0 0 2

### 2. 아래의 문제를 C or C++으로 풀어주세요.

* 문제 설명 : 2차원 평면 위의 점 N개가 주어진다. 좌표를 x좌표가 증가하는 순으로, x좌표가 같으면 y좌표가 증가하는 순서로 정렬한 다음 출력하는 프로그램을 작성하라. 
  
  
* 입력 설명 : 첫째 줄에 점의 개수 N이 주어진다. 둘째 줄부터 N개의 줄에는 i번점의 위치 Xi와 Yi가 주어진다. Xi와 Yi좌표는 항상 정수이고, 위치가 같은 두 점은 없다.
 
  
* 출력 설명 : 첫째 줄부터 N개의 줄에 점을 정렬한 결과를 출력한다.
  
  
* 입력 예시 :
  
5
  
3 4
  
1 1
  
1 -1
  
2 2
  
3 3


* 출력 예시 : 
  
1 -1
  
1 1
  
2 2
  
3 3
  
3 4


### 3. 아래의 문제를 C or C++으로 풀어주세요.

* 문제 설명 : 재민이는 자신의 결혼식에 자신의 학교 친구와 친구의 친구를 초대하기로 했다. 재민이의 학번은 1이며, 사람들간의 친구 관계를 조사한 리스트가 학번으로 주어졌을 때 결혼식에 초대할 사람의 수를 구하라.
  
  
* 입력 설명 : 첫째 줄에 리스트의 길이 N이 주어진다. 
다음 줄부터 N개 줄에는 친구 관계 Xi, Yi가 주어진다. Xi와 Yi는 친구라는 뜻이다.

  
* 출력 설명 : 첫째 줄에 재민이의 결혼식에 초대할 친구의 수를 출력한다.
  
  
* 입력 예시 :
  
5
  
1 2
  
1 3
  
3 4
  
2 3
  
4 5


* 출력 예시 :
  
3
